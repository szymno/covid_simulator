```python
from covid_engine import Engine
from animation import Animation
```

Diseases these days, despite the developed health care, can be very deadly. Why is that? Extensive researches suggest that one of the most important factor is the population density. But is it true? It can be tested (of course theoretically) by using simulations.

Assuming that the simulation area is 200 by 200 units, with 1000 agents of which 100 are infected, 
(other default variables are: 
    speed of an agent = 20 units per step
    incubation time = 14 steps
    convalescense time = 20 steps
    infection probability = 0.25
    infection distance = 2 units
    death probability = 0.02)
results are as follows:

engine = Engine((200, 200), 3000, 100)

engine.place_agents()

animation = Animation(engine)

animation.animate()

As seen, natural immunity was achieved quite fast, in around 40 steps, yet high population density here, as seen, results in high contagiousness and it can lead to problems such as an overload of hospitals and an collapse of the market as many infected need to be quarantined and cannot work. How can we make it smaller? Reducing the overall population density by making the are larger results in:

engine = Engine((300, 300), 3000, 100)

engine.place_agents()

animation = Animation(engine)

animation.animate()

Contagiousness is much smaller, therefore high peak of infected seen on previous chart is also much lower and problems  mentioned before are not as bad as before. It is still not the best situation here, so other variable that can simulate  reduced contacts of susceptible agents with the infected ones is reducing the speed:

engine = Engine((200, 200), 3000, 100, speed=5)

engine.place_agents()

animation = Animation(engine)

animation.animate()

Comparing to the first graph, it also reduced the contagiousness. But what about the face masks? Assuming that every agent wears it, we can reduce the infection spread distance to simulate that:

This not only reduced the contagiousness, it also made the peaks much smaller. Another interesting thing is that many of the agents never were infected. This is very good news for people that are more vulnerable (such as elder, sick etc.). Now let's connect all of these factors:

engine = Engine((300, 300), 3000, 100, infection_distance=1, speed=5)

engine.place_agents()

animation = Animation(engine, steps=125)

animation.animate()

This reduced the disease spread to almost zero, the epidemic died before it could infect the population. This is of course just a simulation and to make this happen, the restrictions would need to be kept at all times. This of course is not possible.

Summarizing, simulations clearly show that factors such as wearing masks in public, reducing the population density at the most crowded places and limiting travel lead to much milder consequences of epidemic, such as the one we happen to experience, so everybody should remember about that.
